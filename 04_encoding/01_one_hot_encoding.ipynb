# One-Hot Encoding (OHE)

## Introduction

Machine learning models require numerical input features. However, many real-world datasets contain categorical variables such as gender, city, or product type.

One-Hot Encoding (OHE) is a feature engineering technique used to convert categorical variables into a numerical format by creating separate binary columns for each category.

Each category is represented as:

- 1 → if the observation belongs to that category  
- 0 → otherwise  

This transformation allows categorical data to be used in machine learning algorithms without introducing ordinal relationships.

---

## Why Use One-Hot Encoding?

- Prevents false ordinal relationships  
- Works well with linear models  
- Makes categorical data model-ready  
- Easy to interpret  

---

## Example

Original Feature:

| City |
|------|
| Pune |
| Mumbai |
| Delhi |

After One-Hot Encoding:

| City_Pune | City_Mumbai | City_Delhi |
|------------|-------------|-------------|
| 1 | 0 | 0 |
| 0 | 1 | 0 |
| 0 | 0 | 1 |
