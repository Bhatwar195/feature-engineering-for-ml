# Label Encoding and Ordinal Encoding

## Introduction

Categorical variables must be converted into numerical format before being used in machine learning models. Label Encoding and Ordinal Encoding are techniques used to transform categorical values into integer representations.

While both methods assign numerical values to categories, they are used in different contexts depending on whether the categories have an inherent order.

---

## Label Encoding

Label Encoding assigns a unique integer to each category.

Example:

| Category | Encoded |
|----------|----------|
| Red      | 0        |
| Blue     | 1        |
| Green    | 2        |

⚠️ Important: Label Encoding does NOT imply order unless the data is truly ordinal.

---

## Ordinal Encoding

Ordinal Encoding is used when categories have a meaningful order.

Example:

| Education Level | Encoded |
|------------------|----------|
| High School      | 0        |
| Bachelor's       | 1        |
| Master's         | 2        |
| PhD              | 3        |

Here, the numerical order reflects the actual hierarchy.
