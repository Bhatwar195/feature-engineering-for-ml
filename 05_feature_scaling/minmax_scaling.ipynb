# Min-Max Scaling (Normalization)

## Introduction

Min-Max Scaling, also known as **Normalization**, is a feature scaling technique that transforms numerical features into a fixed range, usually between **0 and 1**.

This method rescales the data so that the minimum value becomes 0 and the maximum value becomes 1. It preserves the relationships between original data values while adjusting the scale.

---

## Why Min-Max Scaling is Important

- Ensures all features are within the same range  
- Prevents large-value features from dominating the model  
- Improves performance of distance-based algorithms  
- Helps gradient-based optimization converge faster  

---

## Mathematical Formula

Min-Max Scaling transforms a value \(X\) using the formula:

\[
X_{scaled} = \frac{X - X_{min}}{X_{max} - X_{min}}
\]

Where:

- \(X\) = original value  
- \(X_{min}\) = minimum value of the feature  
- \(X_{max}\) = maximum value of the feature  

After transformation:

- Minimum value → 0  
- Maximum value → 1

### Import requirements 

````python
import pandas as pd 
import numpy as np 
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
````

### Load Datset 

````python
df=pd.read_csv("Titanic.csv")
````

````python
df.head()
````

````python
df=df[['Survived','Pclass','Parch']]
df.head()
````

````python
df.isnull().sum()
````

### Train test split 

````python
X=df.drop('Survived',axis=1)
y=df['Survived']
````

````python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
````

````python
# print Shape
print("shape of X_train:",X_train.shape)
print("shape of X_test:",X_test.shape)
print("shape of y_train:",y_train.shape)
print("shape of y_test:",y_test.shape)
````

### Apply min-max scaler  

````python
mx=MinMaxScaler()
````

````python
mx.fit(X_train)
````

````python
X_train_sc=mx.transform(X_train)
X_test_sc=mx.transform(X_test)
````
